```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
header = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'}
```

```python
def soup_(url):
    page = requests.get(url, headers = header)
    soup = BeautifulSoup(page.content,'html.parser')
    soup.prettify()
    return soup
```

```python
soup = soup_("https://www.countryzipcode.com/lithuania")
```

```python
a = soup.find_all('div',{'class':'regions'})
b = a[0].find_all('a')
```

```python
links = []
state = []
for _ in b:
    st = _.get_text()
    li = _['href']
    state.append(st)
    links.append(li)
links, state
```

```python
zipcode = []
states = []
city = []
city_links = []
for i in range(len(links)):
    soup = soup_(links[i])
    a = soup.find_all('div',{'class':'regions'})
    b = a[0].find_all('a')
    for _ in b:
        st = _.get_text()
        print(st)
        li = _['href']
        city_links.append(li)
        for _ in city_links:
            soup = soup_(city_links[i])
            a = soup.find_all('div',{'class':'regions'})
            b = a[0].find_all('a')
            for _ in b:
                st = _.get_text()
                data = st.split(' - ')
                zipcode.append(data[0])
                city.append(data[1])
                states.append(state[i])
                
# [zipcode, states, city]
```

```python
df = pd.DataFrame({
    'state': states,
    'city': city,
    'zipcode': zipcode
})
df.set_index('state', inplace = True)
df.sort_index(inplace = True)
df.drop_duplicates(keep="first", inplace = True)
df
```
